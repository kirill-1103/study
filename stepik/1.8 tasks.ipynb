```python
my_stat = my_stat.rename(columns={'V1':'session_value',
                       'V2':'group',
                       'V3':'time',
                       'V4':'n_users'})
```

```python
my_stat['session_value'] = my_stat['session_value'].fillna(0)
median = my_stat.query('n_users>=0')['n_users'].median()
my_stat.loc[my_stat['n_users']<0,'n_users'] = median
```

```python
mean_session_value_data = my_stat.groupby('group',as_index=False).agg({'session_value':'mean'}).rename(columns={'session_value':'mean_session_value_data'})
```
