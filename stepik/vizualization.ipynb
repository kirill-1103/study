```python
dota_frame = pd.read_csv('dota_hero_stats.csv')
dota_frame.head(2)
```

```python
dota_frame = dota_frame.filter(items=['id','roles'])
counts = pd.Series([dota_frame.roles[i].count(',')+1 for i in range(len(dota_frame.roles))])#  
dota_frame = dota_frame.assign(count_roles = counts)
dota_frame
```

```python
sns.histplot(data=dota_frame.count_roles,bins=15)
```

```python
dota_frame = pd.read_csv('dota_hero_stats.csv')
dota_frame = dota_frame.filter(items=['id','roles'])
sns.histplot([x.count(',')+1 for x in dota_frame.roles],bins=14)
```
