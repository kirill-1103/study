## Группировка groupby

```python
df = sp.groupby('gender').aggregate({'reading_score':'mean','writing_score':'mean','math_score':'mean'})
df = df.assign(s = (df['reading_score']+df['writing_score']+df['math_score'])/3)
df
```

```python
df = sp.groupby('gender').aggregate({'reading_score':'mean','writing_score':'mean','math_score':'mean'})
df['mean_score']=(df['reading_score']+df['writing_score']+df['math_score'])/3
df
```

```python
sp.groupby('gender').mean()
```

```python
sp.groupby('gender', as_index=False)\
    .aggregate({'math_score':'mean','reading_score':'max'})\
    .rename(columns=  {'math_score':'mean_math_score','reading_score':'mean_reading_score'})
```

```python
sp.groupby(['race/ethnicity','gender'],as_index=False).mean()
```

```python
sp.iloc[:,[0,-1,-2,-3]].sort_values(['gender','math_score'],ascending=False).groupby('gender').head()
```

```python
sp2 = sp.assign(total_score = sp.math_score+sp.reading_score+sp.writing_score, mean_score=(sp.math_score+sp.reading_score+sp.writing_score)/3)
sp2.sort_values(['total_score'],ascending=False)
```

# DOTA2

```python
dota_frame = pd.read_csv('dota_hero_stats.csv')
```

```python
dota_frame.groupby('legs').aggregate({'id':'count'})
```

```python
dota_frame.sort_values('legs',ascending=True).head(10)
```

```python
dota_frame[['id','legs']].groupby('legs').legs.count()
```

```python
dota_frame['legs'].value_counts()
```

```python
dota_frame.drop(['Unnamed: 0','id','legs','name','roles'],axis=1)\
    .groupby(['attack_type','primary_attr'],as_index=False)\
    .count()\
    .sort_values('localized_name',ascending=False)\
    .head(1)
```

```python
dota_frame.filter(items=['attack_type','primary_attr']).value_counts()
```
