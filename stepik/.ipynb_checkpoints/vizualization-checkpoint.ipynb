```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sp = pd.read_csv('StudentsPerformance.csv')
sp.head(3)
```

```python
sp = sp.rename(columns = {'math score':'math_score','reading score':'reading_score','writing score':'writing_score'})
sp.math_score.hist()
```

```python
sp.plot.scatter(x='math_score',y='reading_score')
```

```python
sns.lmplot(x='math_score',y='reading_score',data=sp)
```

```python
sns.lmplot(x='math_score',y='reading_score',data=sp,hue='gender')
```

```python
ax =sns.lmplot(x='math_score',y='reading_score',data=sp,hue='gender',fit_reg=False) #  
ax.set_xlabels('Math score')
```
